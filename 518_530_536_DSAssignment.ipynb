```python
# Count null values in each column
null_counts = df.isnull().sum()

# Display null counts
print(null_counts)
```

```python
# Count null values per column
null_counts = df.isnull().sum()
print("Null values in each column:\n")
print(null_counts)

# Show columns with missing values only
print("\nColumns with missing values:\n")
print(null_counts[null_counts > 0])
```

```python
# Calculate percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nPercentage of missing values per column:\n")
print(missing_percentage[missing_percentage > 0])
```

```python
# Fill numeric columns with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())  # assign back instead of inplace

# Fill categorical columns with 'Unknown'
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')  # assign back instead of inplace

# Verify no nulls left
print("\nNull values after handling:\n")
print(df.isnull().sum())
```

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns in dataset:\n", categorical_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode AIRLINE column
le = LabelEncoder()
df['AIRLINE_Label'] = le.fit_transform(df['AIRLINE'])

# Show mapping
mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label Encoding mapping for AIRLINE:\n", mapping)
```

```python
# Take a small sample for demonstration
sample_df = df[['ORIGIN', 'DEST']].sample(5000, random_state=42)

# One-hot encode sample
sample_encoded = pd.get_dummies(
    sample_df,
    columns=['ORIGIN', 'DEST'],
    prefix=['Origin', 'Dest']
)

print(sample_encoded.head())
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Define order including 'Unknown'
ordinal_order = [['Unknown', 'A', 'B', 'C', 'D']]

oe = OrdinalEncoder(categories=ordinal_order)

df['CANCELLATION_CODE_Ordinal'] = oe.fit_transform(
    df[['CANCELLATION_CODE']]
)

print(df[['CANCELLATION_CODE', 'CANCELLATION_CODE_Ordinal']].head())
df[['CANCELLATION_CODE', 'CANCELLATION_CODE_Ordinal']].value_counts()
```

```python
# Frequency encoding for ORIGIN_CITY
freq = df['ORIGIN_CITY'].value_counts() / len(df)
df['ORIGIN_CITY_Freq'] = df['ORIGIN_CITY'].map(freq)

print(df[['ORIGIN_CITY', 'ORIGIN_CITY_Freq']].head())
```

```python
# Create binary delay status feature
df['IS_DELAYED'] = df['DEP_DELAY'].apply(lambda x: 1 if x > 0 else 0)

# Show before and after
print(df[['DEP_DELAY', 'IS_DELAYED']].head())
```

```python
# Create route feature
df['ROUTE'] = df['ORIGIN'] + "_" + df['DEST']

# Show before and after
print(df[['ORIGIN', 'DEST', 'ROUTE']].head())
```

```python
# Create distance categories
df['DISTANCE_CATEGORY'] = pd.cut(
    df['DISTANCE'],
    bins=[0, 500, 1500, 3000, 5000],
    labels=['Short', 'Medium', 'Long', 'Very Long']
)

print(df[['DISTANCE', 'DISTANCE_CATEGORY']].head())
```
